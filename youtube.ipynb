```python
df = df.drop('video_id', axis=1)
df
```

```python
num_cols = ['likes','comments','watch_time_minutes']

for col in num_cols:
    print(f"\nColumn: {col}")
    print("Skewness:", df[col].skew())
```

```python
df['likes'] = df['likes'].fillna(df['likes'].mean())
df['comments'] = df['comments'].fillna(df['comments'].mean())
df['watch_time_minutes'] = df['watch_time_minutes'].fillna(df['watch_time_minutes'].mean())
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix (numerical only)
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features vs Ad Revenue")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Average revenue by Category
cat_revenue = df.groupby('category')['ad_revenue_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
cat_revenue.plot(kind='bar', color='skyblue')
plt.title("Average Ad Revenue by Category")
plt.ylabel("Average Revenue (USD)")
plt.xticks(rotation=45)
plt.show()


# Average revenue by Device
device_revenue = df.groupby('device')['ad_revenue_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,4))
device_revenue.plot(kind='bar', color='orange')
plt.title("Average Ad Revenue by Device")
plt.ylabel("Average Revenue (USD)")
plt.xticks(rotation=45)
plt.show()


# Average revenue by Country
country_revenue = df.groupby('country')['ad_revenue_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
country_revenue.plot(kind='bar', color='green')
plt.title("Average Ad Revenue by Country")
plt.ylabel("Average Revenue (USD)")
plt.xticks(rotation=45)
plt.show()


```
